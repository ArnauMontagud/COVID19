```python
import maboss
```

### Wild type simulation

```python
model = maboss.load("Apoptosis.bnd", "Apoptosis.cfg")
```

```python
res1 = model.run()
res1.plot_piechart()
```

### TNF simulation

```python
model_tnf = model.copy()
model_tnf.network.set_istate(
    'TNF_TNFRSF1A_complex', [0, 1]
)
```

```python
res_tnf = model_tnf.run()
res_tnf.plot_piechart()
```

### FASL simulation

```python
model_fasl = model.copy()
model_fasl.network.set_istate('FAS_FASL_complex', [0, 1])
```

```python
res_fasl = model_fasl.run()
res_fasl.plot_piechart()
```

### TNF + FASL simulation

```python
model_tnf_fasl = model.copy()
model_tnf_fasl.network.set_istate('TNF_TNFRSF1A_complex', [0, 1])
model_tnf_fasl.network.set_istate('FAS_FASL_complex', [0, 1])
```

```python
res_tnf_fasl = model_tnf_fasl.run()
res_tnf_fasl.plot_piechart()
```

```python
model_random = model.copy()
for node in model_random.network.keys():
    model_random.network.set_istate(node, [0.5, 0.5])

model_random.network.set_output([
    'FAS_FASL_complex',
    'BAD_BBC3_BCL2L11_complex',
    'Apoptosome_complex',
    'TNF_TNFRSF1A_complex',
    'TRADD_FADD_complex',
    'BCL2_MCL1_BCL2L1_complex',
    'CASP8',
    'CASP3',
    'CASP9_Cytoplasm',
    'BID',
    'CASP7',
    'Apoptosis_phenotype'
])
```

```python
res_random = model_random.run()
res_random.plot_piechart()
```
